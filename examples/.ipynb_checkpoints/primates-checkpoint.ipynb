```python
import sys, os
sys.path.insert(0,"..")

import numpy as np
import phyloinfer as pinf
```

```python
# load primates data set
data, taxon = pinf.data.loadData('../datasets/primates.nex','nexus')
```

```python
# initialize the parameters
init_newick = '((Bovine:0.3463006,((((Orang:0.1482886,(Gorilla:0.09172454,(Human:0.1151951,Chimp:0.1049027):0.05301964):0.04981972):0.07432181,Gibbon:0.2879228):0.04161085,(BarbMacaq:0.1107167,(CrabEMac:0.1258893,(RhesusMac:0.05893922,JpnMacaq:0.01772492):0.06524793):0.09949257):0.1963516):0.1351803,SquirMonk:0.3171454):0.07502666):0.01274073,(Lemur:0.4248112,Tarsier:0.4240169):0.125704,Mouse:0.5608657);'
init_tree = phmc.Tree(init_newick,format=5)
#init_tree.populate(len(taxon))
#init_tree.unroot()
#phmc.tree.init(init_tree)

print init_tree.get_ascii(attributes=['name','dist'])
```

```python
phmc.tree.namenum(init_tree,taxon)
```

```python
import numpy as np
pden = np.array([0.25, 0.25, 0.25, 0.25])
kappa = 2.0

D, U, beta, rate_matrix = phmc.rateM.decompHKY(pden,kappa)
```

```python
init_branch = phmc.branch.get(init_tree)
L = phmc.Loglikelihood.initialCVL(init_tree,data)
phmc.Loglikelihood.phyloLoglikelihood(init_tree, D, U, beta, pden, L)
```

```python
phmc.infer.Logpost(init_tree, init_branch, D, U, beta, pden, L)
```

```python
sampled_tree, sampled_branch, monitor_U, nni_attempts, accept_rate = phmc.infer.hmc(init_tree, init_branch, pden, kappa, data,
                                                   50, 0.01, 500, surrogate=False, burned=0.2, adap_stepsz_rate=0, delta=0.02)
```

```python
nni_attempts_samp = nni_attempts[100:]
accept_rate_samp = accept_rate[100:]
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(nni_attempts_samp,accept_rate_samp,'.',alpha=0.3)
plt.xlabel('# NNI performed')
plt.ylabel('Acceptance probability')
plt.title(r'$\epsilon=0.005$')
fig.savefig('AP_stepsz_0.005_exact.pdf')
```

```python
def numGoods(numCustomers,lower,upper):
    x = 10000*np.random.rand()
    for i in range(numCustomers):
        action = 20*np.random.rand()-10
        x = max(min(10000,x+action),0)
    return x
```

```python
sample = [numGoods(1000000,0,10000) for i in range(1000)]
```
